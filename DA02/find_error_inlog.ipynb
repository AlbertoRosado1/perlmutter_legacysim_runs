```python
%%time
#https://stackoverflow.com/questions/34530237/find-files-in-a-directory-containing-desired-string-in-python
import os

user_input = "/global/cscratch1/sd/arosado/legacysim/dr9/DA02/north/file0_rs0_skip0/logs"
directory = os.listdir(user_input)

searchstring = "Error"

with open('logs_with_errors.txt', 'w') as new_f:
    for folder in directory:
        #print(f'logs/{folder}')
        folder_path = f'{user_input}/{folder}'
        folder = os.listdir(user_input + os.sep + folder)
        for fname in folder:
            #print(f'{folder_path}/{fname}')
            if os.path.isfile(folder_path + os.sep + fname):
                # Full path
                f = open(folder_path + os.sep + fname, 'r')
                content = f.read()
                if searchstring in content:
                    print(f'found {searchstring} in file {fname}')
                    new_f.write(f'found {searchstring} in file {fname}')
                    new_f.write('\n')
                
                elif searchstring.lower() in content:
                    print(f'found {searchstring} in file {fname}')
                    new_f.write(f'found {searchstring} in file {fname}')
                    new_f.write('\n')
                    
                elif searchstring.upper() in content:
                    print(f'found {searchstring} in file {fname}')
                    new_f.write(f'found {searchstring} in file {fname}')
                    new_f.write('\n')
    
                else:
                    print(f'string not found in file {fname}')
                    new_f.write(f'string not found in file {fname}')
                    new_f.write('\n')
                f.close()
new_f.close()
```
